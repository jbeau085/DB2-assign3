```python
import pandas as pd
from pymongo import MongoClient
from pprint import pprint
client = MongoClient('localhost', 27017)
db = client['newTest']
collection=db['listings']
```

```python
count = 0
for document in collection.find():
    if count < 3:
        print(document)
        count+=1
    else:
        break
```
